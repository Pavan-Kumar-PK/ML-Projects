# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

x0 = np.ones(100).reshape(-1,1)
x1 = np.linspace(-5,5,100).reshape(-1,1)
x2 = x1**2

X = np.concatenate([x0, x1, x2], axis=1)
y = np.dot(X,[1,0.5, 0.8]) + np.random.rand(100)*4
```

```python
#linear regression
coeff = np.dot(np.linalg.inv(np.dot(X.T,X)),np.dot(X.T,y))
print(coeff)
y_pred = np.dot(X, coeff)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(X[:,1],y,s =5,color='red')
plt.plot(X[:,1], y_pred)
plt.show()
```

# Ridge Regression

```python
lm = 0.01
```

```python
# Ridge Regression
coeff_ridge= np.dot(np.linalg.inv(np.dot(X.T,X) + lm*np.identity(3)),np.dot(X.T,y))
```

```python
coeff_ridge
```

```python
y_pred_ridge= np.dot(X, coeff_ridge)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(X[:,1],y,s =5,color='red')
plt.plot(X[:,1], y_pred,'b')
plt.plot(X[:,1], y_pred_ridge, 'g')
plt.show()
```

# Iterative Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

x0 = np.ones(100).reshape(-1,1)
x1 = np.linspace(-5,5,100).reshape(-1,1)
x2 = x1**2
X = np.concatenate([x0, x1, x2], axis=1)
y = np.dot(X,[-5,2, 4]) + np.random.rand(100)*10

# Hyperparameters
learning_rate = 1e-05
num_iterations = 30000

# Initialize coefficients randomly
coefficients = np.random.randn(3)

# Gradient descent loop
for _ in range(num_iterations):
    y_predicted = np.dot(X, coefficients)
    error = y - y_predicted
    gradients = -2 * X.T.dot(error)  # Negative sign for gradient descent
    coefficients -= learning_rate * gradients

# Print the final coefficients
print("Final Coefficients:", coefficients)

original_coefficients = [-5, 2, 4]
print("Original Coefficients:", original_coefficients)

# Visualize
y_predicted = np.dot(X, coefficients)
plt.scatter(x1, y, label="Data")
plt.plot(x1, y_predicted, color='red', label="Prediction")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x0 = np.ones(100).reshape(-1,1)
x1 = np.linspace(-5,5,100).reshape(-1,1)
x2 = x1**2
X = np.concatenate([x0, x1, x2], axis=1)
y = np.dot(X,[-5,2, 4]) + np.random.rand(100)*10

# Hyperparameters
learning_rate = 1e-05
num_iterations = 30000
lambda_reg = 0.1  # Regularization strength

# Initialize coefficients randomly
coefficients = np.random.randn(3)

# Gradient descent loop (Ridge Regression)
for _ in range(num_iterations):
    y_predicted = np.dot(X, coefficients)
    error = y - y_predicted
    gradients = -2 * X.T.dot(error) + 2 * lambda_reg * coefficients  # Ridge modification 
    coefficients -= learning_rate * gradients

# Print the final coefficients
print("Final Coefficients:", coefficients)

original_coefficients = [-5, 2, 4]
print("Original Coefficients:", original_coefficients)

# Visualize
y_predicted = np.dot(X, coefficients)
plt.scatter(x1, y, label="Data")
plt.plot(x1, y_predicted, color='red', label="Prediction")
plt.legend()
plt.show()
```
